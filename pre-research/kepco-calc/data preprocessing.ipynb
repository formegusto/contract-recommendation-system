# Data Processing Process
---
1. Read Data (pd.read_excel, plain data)
2. Convert DataFrame Type
3. peak DataFrame
4. month usage DataFrame

# Read Data

```python
xlsx = pd.read_excel("datas/datas.xlsx", header=None, skiprows=2, engine="openpyxl")
```

# Convert DataFrame Type

```python
# column [7] row [0] - Household Name (아파트명)
# column [7] row [1] - Household Name (동)
# column [7] row [2] - Household Name (호수)
# column [7] row[3] - Start kWh Data 

# column [0] row [3] - YYYYMMDD
# column [1] row [3] - YYYY
# column [2] row [3] - MM
# column [3] row [3] - DD
# column [4] row [3] - hh
# column [5] row [3] - mm

# household list
# dt.datetime(xlsx[1][3], xlsx[2][3], xlsx[3][3], xlsx[4][3], xlsx[5][4])
date_df = xlsx[3:][xlsx.columns[1:6]].copy()
household_df = xlsx[xlsx.columns[7:]]

date_list = [dt.datetime(
    date_df.loc[_][1],
    date_df.loc[_][2],
    date_df.loc[_][3],
    date_df.loc[_][4],
    date_df.loc[_][5]
) for _ in date_df.index]

datas_df = pd.DataFrame(columns=['date'])
datas_df['date'] = date_list

for col in household_df:
    household_name = "{}-{}-{}".format(
        household_df[col][0],
        household_df[col][1],
        household_df[col][2]
    )
    datas_df[household_name] = household_df[col][3:].to_list()
    
datas_df = datas_df.replace("-", 0)
datas_df
```

# peak DataFrame

```python
sum_df = pd.DataFrame(columns=['date', 'kWh', 'kW'])
sum_df['date'] = date_list
sum_df['kWh'] = datas_df[datas_df.columns.difference(['date'])].sum(axis=1).to_list()
sum_df['kW'] = (sum_df['kWh'] / 0.25).to_list()

sum_df
```

```python
peak_df = pd.DataFrame(columns=['month', 'kW'])

for month in range(1,13):
    peak_df = peak_df.append({
        "month": str(month),
        "kW": sum_df[sum_df['date'].dt.month == month]['kW'].max()
    }, ignore_index=True)
    
peak_df
```

# month usage DataFrame

```python
month_usage_df = pd.DataFrame(columns=['month'])
month_usage_df['month'] = [_ for _ in range(1,13)]

for name in datas_df[datas_df.columns.difference(['date'])]:
    self_household_df = datas_df[['date',name]].copy()
    
    month_usage_df[name] = [
        self_household_df[self_household_df['date'].dt.month == month][name].sum()
        for month in range(1,13)
    ]   
    
month_usage_df
```
